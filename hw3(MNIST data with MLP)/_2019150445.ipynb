# 1. Data Loading & Preprocessing

```python
from sklearn.datasets import load_breast_cancer
import tensorflow as tf
import numpy as np
```

```python
X, y = load_breast_cancer(return_X_y=True, as_frame=True)
```

load_breast_cancer 함수를 통해 breast_cancer data를 X, y에 각각 dataframe 형식으로 저장하였다.

```python
print(X.shape)
```

X는 569 instances와 30개의 특성변수를 가지고 있다.

```python
print(y.shape)
```

```python
unique, counts = np.unique(y, return_counts=True)
print('Labels:',dict(zip(unique,counts)))
```

y는 0과 1의 label을 각각 212개, 357개 가지고 있다.

```python
X.head()
```

head를 통해 X의 5개 원소를 찍어보면 다음과 같다. mean area같은 특성변수의 값은 1000정도인데 smoothness eroor같은 특성변수의 값은 0.001즈음인 것을 봐서는 scale의 차이가 있는 것 같다.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, test_size=0.3, stratify=y)
```

train_test_split을 통해 random_state를 주고 y의 분포를 유지하기 위해 stratify를 시키며 test set을 30% 분리하였다.

```python
print(X_train.shape)
print(X_test.shape)
```

# 2. Model selection

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.utils import plot_model

model=Sequential()
model.add(Dense(256, input_dim=30, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))
model.summary()
plot_model(model,show_shapes=True)
```

Input dim은 특성변수가 30개이기에 30으로 잡았다.

첫 번째 Dense층을 dimension 256, activation function은 relu로 적합시켰다.

Overfitting을 방지하기 위해 Dropout을 줘서 30%의 확률로 각각의 노드가 0의 값을 가지게 하였다. 

두 번째 Dense층은 dimension 128, activation function은 위와 같이 relu를 주었고, 모수가 32000개로 데이터에 비해 많음으로 Dropout 0.4를 주었다.

세 번째 Dense층은 dimension 64, activation function은 마찬가지로 relu, Dropout은 0.2를 주었다.

마지막 Dense층은 y의의 label이 1인 확률을 추출해야하기에 dimension=1, activation='sigmoid'로 주었다.

Bottleneck 현상을 막기 위해 dimension이 256부터 점점 작아져 64 dimension에서 최종적으로 한 개의 노드를 추정하게 하였다.


```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
callback_list=[EarlyStopping(monitor='val_accuracy',patience=10),ModelCheckpoint(filepath='model.h5', monitor='val_accuracy',save_best_only=True)]
```

callback을 주기위해 earlystopping, modelcheckpoint를 불러왔고, overfitting을 막기위해 validation_accuracy가 10개의 epoch동안 나아지지 않으면 학습을 중지하도록 하였고, val_accuracy가 개선되면 model을 저장하도록 하였다.

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
history=model.fit(X_train, y_train, epochs=50, batch_size=16, callbacks=callback_list, validation_split=0.3)
history_out=history.history
history_out.keys()
```

model compile 부분을 보면, 추정해야하는 y가 0 또는 1이기에 loss='binary_crossentropy'를 사용하였고, optimizer은 learning rate를 각 parameter 별 다른 크기의 update를 하는 adam을 사용하였다. metric='accuracy'를 설정해 accuracy도 출력되게 하였다.

model train 부분에서는 epochs는 callback에 의해 자동으로 멈추기 때문에 50으로 넉넉히 잡았고, batch_size=16, train set안에서 validation_split을 통해 30%만큼 validation에 이용되게 하였다. 11 epoch에서 학습을 멈췄고, train set의 accuracy는 0.75정도인 반면 validation accuracy는 0.92정도인 것으로 봐서 overfitting은 발생하지 않은 것을 알 수 있다. 

train 과정을 history로 저장해 loss, accuracy, validation loss, validation accuracy를 각 epoch 마다 추출하였다.

```python
model.evaluate(X_test, y_test)
```

test set에 evaluate해보면 0.91정도의 성능으로 괜찮은 성능이면서 overfitting은 발생하지 않은 것을 알 수 있다.

```python
import matplotlib.pyplot as plt

loss=history_out['loss']
loss_val=history_out['val_loss']
accuracy=history_out['accuracy']
accuracy_val=history_out['val_accuracy']
plt.plot(loss,'bo',label='training loss')
plt.plot( loss_val,'b', label='validation loss' )
plt.title('Training and validation losss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

아까 저장한 history를 가지고 train loss와 validation loss를 plot하였다. epoch이 반복되면서 train loss는 확실히 감소하며 수렴하는 양상을 보이는데 validation loss는 처음에 약간 진동하다가 마지막에 와서 수렴하고 있다.

```python
import matplotlib.pyplot as plt

plt.clf()
plt.plot(accuracy,'bo',label='training accuracy')
plt.plot(accuracy_val,'b', label='validation accuracy' )
plt.title('Training and validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

accuracy를 plot하여봐도 마찬가지의 양상을 보인다.

이제 Model을 약간씩 변형할 것이다. 일단 각 층마다 batchnormalization을 추가해 각 배치 별 정규화를 하게 하여 local minimum에 빠질 가능성도 줄이며 learning rate의 영향을 최소화 하였다. 나머지는 위와 같다.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, BatchNormalization
from tensorflow.keras.utils import plot_model

model=Sequential()
model.add(Dense(256, input_dim=30, activation='relu'))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.4))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
model.summary()
plot_model(model,show_shapes=True)
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
history=model.fit(X_train, y_train, epochs=50, batch_size=16, callbacks=callback_list, validation_split=0.3)
history_out=history.history
history_out.keys()
```

학습이 매우 잘 된 것을 볼 수 있다. validation accuracy 0.98까지 올라갔다.

```python
model.evaluate(X_test, y_test)
```

test set에 evaluate해도 overfitting이 발생하지 않으면서 좋은 성능이 나온 것을 확인할 수 있다.

```python
import matplotlib.pyplot as plt

loss=history_out['loss']
loss_val=history_out['val_loss']
accuracy=history_out['accuracy']
accuracy_val=history_out['val_accuracy']
plt.plot(loss,'bo',label='training loss')
plt.plot( loss_val,'b', label='validation loss' )
plt.title('Training and validation losss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

loss를 plot하면 18번째 epoch정도에서 가장 좋은 성능을 보인 것을 알 수 있다. training loss도 validation loss와 같이 커지는 것을 알 수 있다.

```python
import matplotlib.pyplot as plt

plt.clf()
plt.plot(accuracy,'bo',label='training accuracy')
plt.plot(accuracy_val,'b', label='validation accuracy' )
plt.title('Training and validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

마찬가지로 18번쨰 epoch에서 가장 큰 validation, train acc를 가진다. callback에 의해 자동으로 저장되었을 것이다.

마지막으로, model은 위와 같지만 complie 부분에서 learning rate를 약간 건드린 모델이다.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout, BatchNormalization
from tensorflow.keras.utils import plot_model

model=Sequential()
model.add(Dense(256, input_dim=30, activation='relu'))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.4))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization(momentum=0.9))
model.add(Dropout(0.2))
model.add(Dense(1,activation='sigmoid'))
model.summary()
plot_model(model,show_shapes=True)
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
callback_list=[EarlyStopping(monitor='val_accuracy',patience=10),
               ModelCheckpoint(filepath='model.h5', monitor='val_accuracy',save_best_only=True),
               ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3)]
```

ReduceLROnPlateau 함수를 통해 validation_loss가 3 epoch동안 나아지지 않으면 0.3정도로 learning rate를 줄이게 설정하였다. learning rate를 줄임으로써 global minimum에 수렴할 수 있도록 하기 위함이다.

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
history=model.fit(X_train, y_train, epochs=50, batch_size=16, callbacks=callback_list, validation_split=0.3)
history_out=history.history
history_out.keys()
```

learning rate가 6번째 epoch, 12번째 epoch때 30%로 줄어든 것을 알 수 있다. 

```python
model.evaluate(X_test, y_test)
```

이를 test set에 evaluate 했을 때, 0.965정도의 accuracy로 지금까지의 모델 중 가장 좋은 성능을 보인다.
